### Setting up the dataset for predictive modeling using multi-layer perceptron model of a neural network

```python
#Create x and y variables
X = data.drop('Class',axis=1).to_numpy()
y = data['Class'].to_numpy()

#Create Train and Test datasets
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,test_size = 0.20,random_state=100)

#Scale the data
from sklearn.preprocessing import StandardScaler  
sc = StandardScaler()  
x_train2 = sc.fit_transform(X_train)
x_test2 = sc.transform(X_test)
```

```python
#Script for Neural Network
from sklearn.neural_network import MLPClassifier  
mlp = MLPClassifier(hidden_layer_sizes=(9,4,2),
                    activation='relu',solver='adam',
                    max_iter=10000,random_state=100)  
mlp.fit(x_train2, y_train) 
predictions = mlp.predict(x_test2) 

#Evaluation Report and Matrix
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test,predictions))  
print(classification_report(y_test,predictions)) 
```

 from the confusion matrix it can be noted that there are 3+1=4 total wrong predictions out of 137 test data cancer class("2" for benign, "4" for malignant) using multi-layer perceptron model of a neural network

```python
#Forecast Table
predict2 = predictions.T
diff = predict2-y_test
FcstTble=pd.DataFrame({'Actual':y_test,'Predicted':predict2.round(2),'Difference':diff.round(2)})
print('\nForecast Table')
FcstTble.head(10)
```

```python
# Finding the predictions that went wrong in the Class feature using the multi-layer perceptron model of a neural network
indices = [i for i in range(len(y_test)) if y_test[i] != predictions[i]]
wrong_predictions = data.iloc[indices,:]
wrong_predictions
```

### Method to see if performance of model will increase using various strategies.

```python
#Method 1: Increase Hidden Layers
#Script for Neural Network
from sklearn.neural_network import MLPClassifier  
mlp = MLPClassifier(hidden_layer_sizes=(9,5,2),
                    activation='relu',solver='adam',
                    max_iter=10000,random_state=100)  
mlp.fit(x_train2, y_train) 
predictions = mlp.predict(x_test2) 

#Evaluation Report and Matrix
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test,predictions))  
print(classification_report(y_test,predictions)) 
```

### Classification and Regression Trees or CART for short refers to Decision Tree algorithms that can be used for classification or regression predictive modeling 

```python
#Script for Decision Tree
from sklearn.tree import DecisionTreeClassifier  

for name,method in [('DT', DecisionTreeClassifier(random_state=100))]: 
    method.fit(x_train2,y_train)
    predict = method.predict(x_test2)
    print('\nEstimator: {}'.format(name)) 
    print(confusion_matrix(y_test,predict))  
    print(classification_report(y_test,predict)) 
```

### b.State and Explain Final Algorithm 

Multi Layer Perceptron Model


The multilayer perceptron model is a type of neural networking model that matches relevant inputs with related outputs. It is comprised of numerous layers of nodes.  The input node is different from  the other  nodes and can be thought of as neuron, or processing core, with a specific activation function.. The multilayer perceptron model is capable of distinguishing data that is not separable in a linear way. The multilayer perceptron model uses the uses backpropagation algorithm that can be seen in a standard neural network. Multilayer perceptron’s can be found as part of the standard analysis algorithm for pattern recognition processes.


A perceptron has one or more inputs, a bias, an activation function, and a single output. In this cancer detection model, The perceptron receives 9 inputs, multiplies them by some weight, and then passes them into an activation function to produce an output. An activation function is a simple function that changes inputs into outputs that have a certain range. The activation used in this model is known as “ReLu” or Rectified Linear Unit. ReLu is a non-linear activation function that is used in multi-layer neural networks or deep neural networks. This function can be represented as:





The output of ReLu is the maximum value between zero and the input value.An output is equal to zero when the input value is negative and the input value when the input is positive. 




Bias is also added to the perceptron, which prevents all  inputs from becoming equal  to zero.

There are 4 hidden layers of nodes, and 2 classes categorized as “2” for a benign cancer and “4” for a “malignant” cancer.
The model took a maximum of 10000 iterations, with a train size of 0.8 and a test size of 0.2 to run the model on.

## 3.Identify 3 ways to Improve Performance of Model

1.Feature Selection-removing features in order to improve model performance. 
feature selection can be done  on a neural network  through creating a decision tree model.
Decision trees are a separate type of machine learning model. Decision trees are useful as they can calculate and rank the most important featuresbased on various statistical criteria. We can first fit the data using the decision tree, and
then use the output from the decision tree to remove features that are deemed unimportant,
before providing the reduced dataset to our neural network.


2.Regularization - is a  technique which helps to reduce overfitting in a neural network. When we add regularization to  network, we add a new regularization term, and the loss function is modified.Dropout regularization is another regularization technique in which certain neurons are dropped  along with their connections present in the neural network.
Dropout regularization prohibits the neurons from relying only on certain features, and weights are spread out amongst all the inputs. The network uses all the input features to determine the output and does not rely on any single one, thus improving the neural network model.

Ensemble of Algorithms-If individual neural networks are not as accurate as benchmark that is ideal, method of emsemble is when various neural networks are combined to increase predictive power. Different neural network architectures can be used, and train the various neural networks on  on different parts of the data and ensemble them and use their collective predictive power to get high accuracy on test data.
